2.Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]


```python
#Calculando as medidas estatísticas, conforme solicitado:

#Criando uma tabela dinamica para calculcar a media, moda e mediana das horas de sono por profissao

df.medidas_estatisticas = pd.pivot_table(df, values='Duração do sono', index='Profissao', aggfunc=[np.mean, pd.Series.mode, np.median], fill_value=None)

df.medidas_estatisticas.columns = ['Media','Moda','Mediana']

df.medidas_estatisticas['Media'] = df.medidas_estatisticas['Media'].round(2)

df.medidas_estatisticas
```

4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)


```python
# Utilizando o metodo "isin" para filtrar as linhas correspondentes as profissões de advogado e representante de vendas
# Lista das profissões correspondentes
profissoes_correspondentes = ['Advogado(a)', 'Representante de Vendas']

# Filtrando o DF apenas com as profissões correspondentes com o método isin
df_profissoes_correspondentes = df[df['Profissao'].isin(profissoes_correspondentes)]

# Calcular a média geral da coluna 'Duração do sono'
media_geral = df['Duração do sono'].mean()

# Calculando a media da coluna 'Duração do sono' para adv e  vendas
media_duracao_sono_adv_vendas = df_profissoes_correspondentes.groupby('Profissao')['Duração do sono'].mean()

# Exibe a media da duração do sono para cada
if media_geral < media_duracao_sono_adv_vendas.min():
  print("Advogado(a) ou Representante de Vendas não tem menos horas de sono ")
else:
  print("Advogado(a) ou Representante de Vendas tem menos horas de sono")
```

5.Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)

```python
# Utilizando o metodo "isin" para filtrar as linhas correspondentes as profissões de Enfermeiro e Medico(a)

# Lista das profissões correspondentes
profissoes_correspondentes = ['Enfermeiro(a)', 'Médico(a)']

# Filtrando o DF apenas com as profissões correspondentes com o método isin
df_profissoes_correspondentes = df[df['Profissao'].isin(profissoes_correspondentes)]

# Calculando a media da coluna 'Duração do sono' para Enfermeiro(a) e Medico(a)
media_duracao_sono_enf_med = df_profissoes_correspondentes.groupby('Profissao')['Duração do sono'].mean()

# Verificando quem tem menos horas de sono
profissao_menos_horas_sono = media_duracao_sono_enf_med.idxmin()

# Arredondando para duas casas decimais
media_duracao_sono_enf_med = media_duracao_sono_enf_med .round(2)

# Exibe a media da duração do sono para cada profissão
print("Media da duraçao do sono: ")
for profissao, media in media_duracao_sono_enf_med.to_dict().items():
  print(f"{profissao}:{media:.2f}")

print(f"\nComparando a media de sono de Enfermeiro e Médico(a), quem tem menos horas de sono é : {profissao_menos_horas_sono}")
```

8.Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)


```python
# Tratamento dos valores da coluna Pressão sanguínea separadamente
df[['Pressao Sistolica', 'Pressao Diastolica']] = df['Pressao Sanguinea'].str.split('/', expand=True)

df['Pressao Sistolica'] = pd.to_numeric(df['Pressao Sistolica'], errors='coerce')
df['Pressao Diastolica'] = pd.to_numeric(df['Pressao Diastolica'], errors='coerce')

# Formula padrão para encontrar a media da pressão sanguinea PAM
PAM = (df['Pressao Sistolica'] + 2 * df['Pressao Diastolica']) / 3

df_media_ps = PAM

#Recebe a media da pressão sanguinea-formula padrão para calcular a PAM

df['df_media_ps'] = PAM

#Agrupa as colunas por gênero(sexo)

max_media_ps = df.groupby('Gênero')['df_media_ps'].mean().idxmax()

print('O gènero que apresenta maior pressão sanguinea média é:', max_media_ps)
```

9.É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?

```python
#Verificar a moda de duração do sono

moda_duracao_sono = df['Duração do sono'].mode()

predominancia_sono_8 = 8 in moda_duracao_sono.values

if predominancia_sono_8 :
  print("Sim, é predominante, os participantes dormirem 8h por dia!")

else:
  print("Não é predominante os participantes dormirem 8h por dia!")

```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
#Coluna que indica se a frequência cardiáca pe maior que 70

df['Frequência cardíaca_acima_70'] = df['Frequência cardíaca'] > 70

#Calculando a média de passos para o grupo com frequência acim de 70 e igual ou abaixo de 70
media_passos_acima_70 = df[df['Frequência cardíaca_acima_70']]['Passos diários'].mean()
media_passos_70_ou_menos = df[~df['Frequência cardíaca_acima_70']]['Passos diários'].mean()


# Verificar se a média de passos é maior para frequências acima de 70
if media_passos_acima_70 > media_passos_70_ou_menos:
    print("Sim! Pessoas com frequências cardíacas acima de 70 dão mais passos.")
else:
    print("Não! As pessoas com frequências cardíacas acima de 70 não dão mais passos que pessoas com frequência cardiaca menor ou igual a 70.")
```
